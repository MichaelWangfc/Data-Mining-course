## 绘制不同类型图形

### 散点图 

```python
x = np.arange(10)   #生成数据：点的x轴，y轴坐标
y=  np.random.randint(0,10,size =10)
size = np.random.rand(10) * 50 #随机生成点的大小的随机数
colour = ['red','green','blue','yellow','cyan','black','magenta','#fff0f5','#ffdab9','#ffa500']
plt.scatter(x, y, size, c= colour)  #生成散点图，设置每个点的大小和颜色
plt.title('scatter plot')
plt.show()
```

### 直方图 

```python
mean = 100 # 设置平均值和标准差
sigma = 10
x = mean + sigma * np.random.randn(10000) # 生成10000个均值为100，标准差为10的随机数

# 设置直方图版式：10000个随机数，50个格子，背景色为蓝色，透明图为0.5
n,bins,patches = plt.hist(x, bins =50,facecolor='b', alpha=0.5)

plt.title("Histogram of IQ: $\mu=100$, $\sigma=10$")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid() #生成网格线 
plt.subplots_adjust(left=0.15) #左边距为0.15英寸
plt.savefig("Histogram_demo.png")
plt.show()
```

### 等高线图

```python
def f(x,y):  #自定义函数 
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)  #生成X,Y轴的二维网格坐标

plt.contourf(X, Y, f(X,Y), 8, cmap='jet') #生成等高线图形

plt.show()
```

```python
def f(x,y):  #自定义函数 
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)  #生成X,Y轴的二维网格坐标

plt.subplots(figsize = (12,6))

plt.subplot(121)
plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)

plt.subplot(122)
plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet') #生成等高线图形
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.show()
```

### 多子图

```python
x = np.linspace(0, 2*np.pi, 100) #生成数据
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (8,4))  #生成Figure对象，并设置宽度和高度

plt.subplot(1,2,1) #创建1*2的子图，并在第一个子图上设置颜色和标题
plt.plot(x,y_sin,'r')
plt.title('Sine')

plt.subplot(1,2,2) #生成第二个子图，并设置标题
plt.plot(x,y_cos)
plt.title('Cos')
```

```python
x = np.linspace(0, 2*np.pi, 100) #生成数据
y_sin = np.sin(x)
y_cos = np.cos(x)

figure,axes = plt.subplots(1,2,figsize = (8,4)) #创建1*2的子图，并在第一个子图上设置颜色和标题
axes[0].plot(x,y_sin,'r')
plt.title('Sine')
axes[0].set_xticks(np.linspace(0,2*np.pi,5))
axes[0].set_xticklabels([0,r'$0.5\pi$', r'$\pi$', r'$1.5\pi$', r'$2\pi$'])

# plt.xticklabels([0,r'$0.5\pi$', r'$\pi$', r'$1.5\pi$', r'$2\pi$'])

axes[1].plot(x,y_cos)
plt.title('Cos')
```

### 对数坐标轴

```python
w= np.linspace(0.1,1000,1000)
p =np.abs(1/(1+0.1j*w))#计算低通滤波器的频率响应
plt.subplot(121)  #生成原始数据的图形
plt.plot(w,p)
plt.ylim(0,1.5)
plt.title('original graph')

plt.subplot(122) #生成x轴为对数的图形
plt.semilogx(w,p)
plt.ylim(0,1.5)
plt.title('graph with logx')
```

### 极坐标轴

```python
theta = np.arange(0,2*np.pi,0.02)  #设定theta的取值范围
plt.subplot(polar=True)      #生成图形画布和子图

plt.plot(3*theta,theta/3,'--')  # 绘制子图

plt.rgrids(np.arange(0.5,2,0.5),angle =45) #设置同心圆栅格的半径大小和文字标注的角度
plt.title(r'r=$\theta$/3,$\alpha$ = 3*$\theta$') #设置子图标题
```

### 绘制3D图形


```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure() #生成当前画布
ax = Axes3D(fig)# 生成当前3D子图
X = np.arange(-4, 4, 0.25) #生成一维数据的范围
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y) #生成二维网格数据的（X,Y）
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot') #使用plot_suface方法，绘制3D曲面

plt.show()
```

```python
def f(x,y):  #自定义函数 
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)  #生成X,Y轴的二维网格坐标


fig = plt.figure()      #生成画布实例

ax = fig.gca(projection='3d') #生成当前3D子图：3D-aware axes实例

ax.contour(X, Y, f(X,Y), 8, cmap='jet') #具体在 Axes3D生成等高线图形

plt.show()
```

```python
X = np.random.randn(1000, 2)

#创建3D对象
fig, ax = plt.subplots(1, 1, figsize=(8, 6), subplot_kw={'projection': '3d'})
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100) #生成参数范围
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='3D curve') #生成3D曲线图
ax.legend()
 
plt.show()
```

### 绘制3D轮廓

```python
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
fig = plt.figure()
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)
cset = ax.contour(X, Y, Z, zdir='z', offset=-100, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-40, cmap=cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=40, cmap=cm.coolwarm)
 
ax.set_xlabel('X')
ax.set_xlim(-40, 40)
ax.set_ylabel('Y')
ax.set_ylim(-40, 40)
ax.set_zlabel('Z')
ax.set_zlim(-100, 100)
 
plt.show()
```

### 绘制3D网格线

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.05)
 
# Plot a basic wireframe.
ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)
```
