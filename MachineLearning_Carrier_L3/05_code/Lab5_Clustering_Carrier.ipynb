## 业务场景
本案例使用电信用户的通信行为数据集，进行用户信用分群和分析。由于是没有标注的训练样本，使用降维和聚类等无监督方法将用户进行分群。<br>
然后对不同群体数据进行人工分析，确定群体的信用行为特点，给用户画像。进而可以运用到后期如精准营销等应用中。

## step1: 数据读取

```python
#导入常用的库
# numpy: python中常用的多维数值计算库，常用简称 np
# pandas:python中的数据分析库，常用简称 pd
# matplotlib :python中的图形库，常用简称plt
# %matplotlib inline：IPython魔法函数（Magic Functions），可以内嵌绘图，并且可以省略掉plt.show()这一步
#约4行代码：

import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
#设定绘图时的中文字体的设置
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

## 读取运营商数据

```python
#使用pandas的读取csv文件的函数 pd.read_csv(),生成数据的dataframe
#数据文件名：telecom.csv
#编码格式：utf8
#分隔符： ,
#约1行代码:

X = pd.read_csv('../data/telecom.csv',encoding='utf-8')

#显示前5行:
#约1行代码:
X.head()

#显示数据的形状大小
#约1行代码:
print(X.shape)
```

## 数据标准化

```python
#导入数据预处理包
from sklearn import preprocessing

#进行数据标准化
# 提示：使用preprocessing.scale()方法
#约1行代码:
X_scaled = preprocessing.scale(X)

#显示标准化后数据
#约1行代码:
print(X_scaled[0:5])
```

```python
#验证标准化后数据的平均值
# 提示：使用np.mean()函数
#约1行代码:
print(np.mean(X_scaled,axis=0))

#验证标准化后数据的标准差
# 提示：使用np.std()函数
#约1行代码:
print(np.std(X_scaled,axis=0))
```

## 进行PCA数据降维

```python
#导入sklearn.decomposition类中的PCA()方法
#约1行代码:
from sklearn.decomposition import PCA

#生成PCA实例
#提示：参数n_components表示生成的主成分个数，我们设置n_components=3
#约1行代码:
pca = PCA(n_components=3)

#在训练集上进行PCA降维转换
#提示:使用PCA.fit_transform()方法
#约1行代码:
X_pca = pca.fit_transform(X_scaled)

#生成降维后的dataframe
# 提示：使用pd.DataFrame()方法，列名设置为columns=['pca_1','pca_2','pca_3']
# 约1行代码:
X_pca_frame = pd.DataFrame(X_pca,columns=['pca_1','pca_2','pca_3'])

#显示三个主成分构成的dataframe的前5行
# 约1行代码:
X_pca_frame.head()
```

```python
#显示各主成分的方差
# 提示：使用pca.explained_variance_属性
#约1行代码:
print(pca.explained_variance_)

#显示各主成分的方差比例
# 提示：使用pca.explained_variance_ratio_ 属性
#约1行代码:
print(pca.explained_variance_ratio_)
```

```python
#查看这三个维度坐标轴（主成分）与原始的7个维度坐标轴之间的关系：
#提示：使用 pca.components_每行表示主成分与原来特征的系数，列名columns = X.columns,索引值 index =['pca_1','pca_2','pca_3']
#约1行代码:
pd.DataFrame(pca.components_,columns = X.columns,index =['pca_1','pca_2','pca_3'])
```

##  K-means 聚类

### 训练简单模型

```python
#导入KMeans的算法包
#提示：使用sklearn.cluster类中的 KMeans方法
#约1行代码:
from sklearn.cluster import KMeans

#KMeans算法实例化
#提示：可以根据自己数据实际情况设置K值，这儿将其设置为K=10
#约1行代码:
est = KMeans(n_clusters=10,random_state=117)

#作用到降维后的数据上
#约1行代码:
est.fit(X_pca)
```

```python
#取出聚类后的标签
#提示：使用KMeans.labels_属性
#约1行代码:
kmeans_clustering_labels = pd.DataFrame(est.labels_,columns=['cluster'])

#将聚类的标签作为新的一列添加到原降维后的矩阵，生成有聚类后dataframe
#提示：使用pd.concat([X,Y],axis=1)
#约1行代码:
X_pca_frame_new = pd.concat([X_pca_frame,kmeans_clustering_labels],axis =1)

#显示新的矩阵的前五行
#约1行代码:
X_pca_frame_new.head()
```

### 对不同的k值进行计算，筛选最优的k值

```python
#导入matplotlib的三维对象
#约1行代码:
from mpl_toolkits.mplot3d import Axes3D

#导入模型测度的类
#约1行代码:
from sklearn import metrics
```

```python
#建立一个空字典，存储每个k对应的评估值 {k：calinski_harabaz_score，...}
#约1行代码:
d = {}

#设置图形大小
fig_reduced_data = plt.figure(figsize= (12,12))

#KMeans的K值设置为K=range（2,14）进行迭代运算
for k in range(2,14):
    #对不同K值进行实例化，random_state=111
    est = KMeans(n_clusters=k,random_state=111)
    
    #作用到降维后的数据上
    y_pred = est.fit_predict(X_pca)
    
    #评估不同k值聚类算法效果
    #提示：使用metrics.calinski_harabaz_score
    score = metrics.calinski_harabaz_score(X_pca_frame,y_pred)
    d.update({k:score})
    print('calinski_harabaz_score with k={0} is {1}'.format(k,score))
    #生成三维图形，每个样本点的坐标分别是三个主成分的值  
    ax= plt.subplot(4,3,k-1,projection = '3d')
    ax.scatter(X_pca_frame.pca_1,X_pca_frame.pca_2,X_pca_frame.pca_3,c= y_pred)
    ax.set_xlabel('pca_1')
    ax.set_ylabel('pca_2')
    ax.set_zlabel('pca_3')
```

```python
#绘制不同k值对应的score，找到最优的k值
x = []
y = []

for k,score in d.items():
    x.append(k)
    y.append(score)   

plt.plot(x,y)
plt.xlabel('k value')
plt.ylabel('calinski_harabaz_score')
```

### 使用最优的k值重新进行聚类

```python
#KMeans算法实例化，random_state=119
#约1行代码:
est = KMeans(n_clusters=12,random_state=119)

#作用到降维后的数据上
#约1行代码:
est.fit(X_pca)

#取出聚类后的标签
#提示：使用KMeans.labels_属性
#约1行代码:
kmeans_clustering_labels = pd.DataFrame(est.labels_,columns=['cluster'])

#将聚类的标签作为新的一列添加到原降维后的矩阵，生成有聚类后dataframe
#提示：使用pd.concat([X,Y],axis=1)
#约1行代码:
X_pca_frame = pd.concat([X_pca_frame,kmeans_clustering_labels],axis =1)

#显示新的矩阵的前五行
#约1行代码:
X_pca_frame.head()
```

### 样本筛选

```python
#合并原数据和三个主成分的聚类后的数据
#约1行代码:
X_full = pd.concat([X,X_pca_frame],axis = 1)

#显示前5行
#约1行代码:
X_full.head()
```

去除异常数据样本：  
通过对聚类数据进行筛选，删除每一个聚类中的噪音和异常点，提取每一个聚类中的典型用户。为了避免离群点的影响，按照每一类的分位数筛选离中心点较近的用户作为该类的代表用户。筛选后的用户样本信用分数分布如图所示。


```python
#按每个聚类分组
#约1行代码:
grouped = X_full.groupby('cluster')

#结果存储到result_data
#约1行代码:
result_data  = pd.DataFrame()

#对分组做循环，分别对每组进行去除异常值处理
#计算去除异常值前的样本数据量
print('Ssample:',X_full['pca_1'].count())
for name,group in grouped:
    #计算每组去除异常值前的个数
    print('Group:{0}, Samples before:{1}'.format(name,group['pca_1'].count()))
    
    #计算每组的统计数据
    desp = group[['pca_1','pca_2','pca_3']].describe()
    
    #对每个特征进行异常值处理：使用1.5IQR法则
    for att in ['pca_1','pca_2','pca_3']:
        #1/4分位值
        lower25 =desp.loc['25%',att]
        #3/4分位值
        upper75 =desp.loc['75%',att]
        #IQR
        IQR = upper75 - lower25
        #最小阈值
        min_value = lower25-1.5*IQR
        #最大阈值
        max_value = upper75 +1.5*IQR
        #使用统计中的1.5*IQR法则，删除每个聚类中的噪音和异常点
        group = group[(group[att]>min_value)&(group[att]<max_value)] 
    #结果存储到result_data 
    result_data = pd.concat([result_data,group],axis = 0)
    
    #计算每组去除异常值后的个数
    print('Group:{0}, Samples after:{1}'.format(name,group['pca_1'].count()))
#计算去除异常值后的数据量
print('Remain sample:',result_data['pca_1'].count())
```
